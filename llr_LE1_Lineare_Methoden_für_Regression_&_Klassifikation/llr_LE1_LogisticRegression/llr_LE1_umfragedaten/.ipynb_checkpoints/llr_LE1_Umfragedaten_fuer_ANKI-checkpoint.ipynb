```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from scipy.stats import norm
```

```python
data = pd.read_csv("umfragedaten_v1_an-1636985768.csv")
data = data.iloc[0:100]
data = data.dropna()
```

```python
data.plot.scatter(x = "GRO", y = "GEW")
plt.show()
```

```python
x = data["GRO"].values.reshape(-1,1)
y = data["GEW"].values.reshape(-1,1)

model = LinearRegression().fit(x,y)

r_sq = model.score(x,y)
steigung = model.coef_
achsenabschnitt = model.intercept_
```

```python
data.plot.scatter(x = "GRO", y = "GEW")
plt.plot(x, steigung * x + achsenabschnitt, color = "r")
plt.show()
```

```python
predict_gew = model.predict(x)

residuen = y - predict_gew

plt.plot(x, residuen, "o", color = "b")
plt.show()
```

```python
n, bins, patches = plt.hist(residuen, 10, stacked = True, density = True)


mue = np.mean(residuen)
sigma = np.std(residuen)

y = norm.pdf(bins, mue, sigma)
plt.plot(bins, y, "r--")
plt.show()
```
