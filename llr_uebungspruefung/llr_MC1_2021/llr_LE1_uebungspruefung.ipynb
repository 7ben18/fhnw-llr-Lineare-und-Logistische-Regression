```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_excel("auto.xlsx")
data = data.set_index("auto")
```

```python
data.plot.scatter(x = "Geschwindigkeit", y = "Bremsweg")
```

```python
x = data["Bremsweg"].values.reshape(-1,1)
y = data["Geschwindigkeit"].values
```

```python
model = LinearRegression().fit(x,y)
```

```python
R2 = model.score(x,y)
steigung = model.coef_
achsenabschnitt = model.intercept_
print("y = ", steigung[0], " * x ", achsenabschnitt)
```

```python
plt.scatter(x, y)
plt.plot(x, model.coef_ * x + model.intercept_, color = "r")
```

```python
y_predict = model.predict(x)
residuen = y - y_predict
```

```python
plt.scatter(x, residuen)
plt.plot(x, x * 0, c = "r")
```

```python
plt.hist(residuen, bins = 20, stacked = True, density = True)
```

```python
x_log = np.log(x)
y_log = np.log(y)
plt.scatter(x_log, y_log)
```
