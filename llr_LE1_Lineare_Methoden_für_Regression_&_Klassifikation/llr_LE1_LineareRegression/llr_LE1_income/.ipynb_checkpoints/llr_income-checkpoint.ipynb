```python
# Module Importieren
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression, LogisticRegression
```

```python
# Daten in Pandas einlesen
data = pd.read_csv("income.data.csv", index_col = "Unnamed: 0")
data.head()
```

```python
# Anzahl Zeilen und Spalten des Dataframes
data.shape
```

```python
# x und y Werte definieren und in Liste von Liste transformieren (x-Achse)
x = data["income"].values.reshape(-1,1)
y = data["happiness"].values
```

```python
# Datenpunkte plotten als Streudiagram
plt.scatter(x, y)
plt.show()
```

```python
# Mein Modell erstellen von einer Linearen Geraden
model = LinearRegression().fit(x,y)
```

```python
# Parameter der Linearen Gerade herausfinden
steigung = model.coef_
achsenabschnitt = model.intercept_
R2 = model.score(x,y)
print("y = ", steigung[0], "* x ", achsenabschnitt)
print("Bestimmtheitsmass: ", R2)
```

```python
# Streupunkte 
plt.scatter(x,y)
plt.plot(x, x * steigung + achsenabschnitt, color = "r")
plt.show()
```

```python
# Residuen berechnen
y_regression = model.predict(x)
residuen = y - y_regression
```

```python
# Residuen Plotten
plt.scatter(x, residuen)
plt.plot(x, 0 * x, color = "r")
plt.show()
```

```python
# Residuenanalyse - Histogramm
plt.hist(residuen, bins = 10, stacked = True, density = True)
plt.show()
```

```python
# x und y Werte Logarithmieren
x_log = np.log(x)

y_log = np.log(y)
```

```python
# log(x) und y Werte plotten
plt.scatter(x_log, y)
plt.show()
```

```python
# Lineare Regression und Bestimmtheitsmass berechnen
model_xlog = LinearRegression().fit(x_log, y)
R2_xlog = model_xlog.score(x_log, y)
print("Bestimmtheitsmass:", R2_xlog)
```

```python
# x und log(y) Werte plotten
plt.scatter(x, y_log)
plt.show()
```

```python
# Lineare Regression und Bestimmtheitsmass berechnen
model_ylog = LinearRegression().fit(x, y_log)
R2_ylog = model_ylog.score(x, y_log)
print("Bestimmtheitsmass:", R2_ylog)
```

```python
# log(x) und log(y) Werte plotten
plt.scatter(x_log, y_log)
plt.show()
```

```python
# Lineare Regression und Bestimmtheitsmass berechnen
model_xlog_ylog = LinearRegression().fit(x_log, y_log)
R2_xlog_ylog = model_xlog_ylog.score(x_log, y_log)
print("Bestimmtheitsmass:", R2_xlog_ylog)
```
